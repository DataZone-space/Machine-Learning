```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split # train_test_split function
from sklearn import metrics # scikit-learn metrics module to calculate accuracy and other metrics
iris = datasets.load_iris()
```

```python
iris.keys()
```

```python
X = iris.data[:, 2:] # petal length and petal width
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
'''Dividing the dataset into a training set that we train the model on and a test set that we evaluate the model's
performance on.'''
```

```python
clf = DecisionTreeClassifier(max_depth=2)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn import tree
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(25,20))
tree.plot_tree(clf,
feature_names=iris.feature_names[2:],
class_names=iris.target_names,
rounded=True,
filled=True);
```
