```python
import numpy as np
from sklearn.linear_model import LinearRegression 
'''The model we are going to use, you usually wouldn't code the linear regression model yourself,
as there are libraries like sci-kit learn which have these models as objects that you can use, we will use a lot of
models from sci-kit learn as it is the most used library in using machine learning models'''
import matplotlib.pyplot as plt
# ^ matplotlib is another library that we will use a lot in the future
np.random.seed(42) 
# We are setting a seed so that the notebook gives the same result every time
```

```python
#let's generate a dataset
X = 3.5 * np.random.rand(370, 1)
# ^ This line generates 370 values of x between 0 and 3.5 * 1, the 1 besides 370 indicates that it is a one
# dimensional vector
y = 5 + 4 * X + np.random.randn(370, 1) 
# ^ This line generates 370 negative and positive values of y with a mean of 0 and variance of 1
X_test = np.array([[0], [3.5]]) 
# ^ This generates two values, 0 and 3.5, which we will use to test and plot the linear regression model
plt.scatter(X, y)
plt.show()
# ^ Plot the X and y values in a scatterplot
```

```python
lin_reg = LinearRegression()
# The linear regression model in sci-kit learn uses the normal equation to find the best parameter values
lin_reg.fit(X, y)
# Fit the model to the data we have
lin_reg.intercept_ 
# The bias, or intercept it estimated
lin_reg.coef_ 
# The parameter value it estimated for the feature, this becomes a vector if there are more than one feature
```

```python
y_predict = lin_reg.predict(X_test)
# The model uses the equation it estimated to predict future values
y_predict
```

```python
plt.plot(X_test, y_predict, "-r"#Make the prediction line red
         , linewidth=2, label="Predictions")
# Plot the line between the two points we predicted, which gives us the regression line
plt.scatter(X, y)
# Plot the line over the previous scatter plot
plt.legend(loc="upper left", fontsize=13)
# Make a legend
plt.axis([0, 3.5, 0, 23])
# Set the limits for the axes
plt.show()
```
