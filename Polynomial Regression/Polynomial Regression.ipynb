```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
np.random.seed(42)
#We are setting a seed so that the notebook gives the same result every time
```

```python
X = 8 * np.random.rand(250, 1) - 4 #np.random.rand generates values between 0 and 1
#This line generates 250 values of x between -4 and (4 * 1), in (250,1), the 1 indicates it is a one-dimensional vector

y = 3 + 3 * X + 1.5 * (X**2) + 0.5 * (X**3) + 2.5* np.random.randn(250, 1)#Generates 250 negative and positive values with a mean of 0 and variance of 1 to add noise
#This line generates 250 values of y
```

```python
plt.scatter(X, y)
plt.xlabel("$X$", fontsize=16)
plt.show()
# Plot the data
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree = 3, include_bias=False)
X_poly = poly_features.fit_transform(X)
```

```python
X[0],X_poly[0]
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
X_test = np.linspace(-4, 4, 100).reshape(100, 1) #Generates 100 random values from 0 to 3
y_test = lin_reg.predict(X_test)#Predicts the outputs of the test data
plt.scatter(X, y)
plt.plot(X_test, y_test, "r-", linewidth=3, label="Predictions")
plt.xlabel("$x_1$", fontsize=16)
plt.ylabel("$y$", rotation=0, fontsize=16)
plt.legend(loc="upper left", fontsize=12)
plt.axis([-4, 4, -20, 80])
plt.show()
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
X_test = np.linspace(-4, 4, 100).reshape(100, 1) #Generates 100 random values from 0 to 3
X_test_poly = poly_features.transform(X_test)#Transforms the new test data to add the new features
y_test = lin_reg.predict(X_test_poly)#Predicts the outputs of the test data
plt.scatter(X, y)
plt.plot(X_test, y_test, "r-", linewidth=3, label="Predictions")
plt.xlabel("$x_1$", fontsize=16)
plt.ylabel("$y$", rotation=0, fontsize=16)
plt.legend(loc="upper left", fontsize=12)
plt.axis([-4, 4, -20, 80])
plt.show()
```
